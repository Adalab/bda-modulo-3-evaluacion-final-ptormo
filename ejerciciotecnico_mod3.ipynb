```python
#posible union 
df_union = pd.merge(df_actividadvuelos, df_historialfidelidad, on='Loyalty Number', how='left') 
#mantener toda la info de la primera, añadiendo el historial de clientes, ya que cada fila de actividad vuelos va a ser combinada con historial fidelidad, pues es la que tiene los loyalty number sin duplicados
df_union.head(5)
```

```python
df_union["Loyalty Number"] =df_union["Loyalty Number"].astype(str) #cambio de loyalty number a string para no tratarlo como número
```

TRATAMIENTO DE NULOS: cancellation year y month

```python
#Transformación de nulos a 0, ya que no tienen año de cancelación porque no han cancelado
df_union['Cancellation Year']= df_union['Cancellation Year'].fillna(0).astype(int) 

```

```python
#Transformación de nulos a 0, ya que no tienen mes de cancelación porque no han cancelado
df_union['Cancellation Month']= df_union['Cancellation Month'].fillna(0).astype(int) 
```

```python
#crear columna booleana si han cancelado: si mes y años son diferentes a 0 han cancelado, después map para decir si sí o si no han cancelado en vez de true o false
df_union['Cancelado']= (df_union['Cancellation Month']!= 0) & (df_union['Cancellation Year']!=0)
```

```python
round(df_union['Cancelado'].value_counts(normalize=True)*100,3) #devuelve el valor entre 0 y 1 y luego *100 para el porcentaje
```

TRATAMIENTO NULOS SALARY

```python
df_union['Salary'].isna().sum()/ df_union.shape[0]*100 #nulos de salario
```

```python
df_union['Education'].unique()
```

```python
df_union.loc[df_union['Education'] == 'College', 'Salary'].isna().sum() #los nulos son solo de college
```

```python
#calculamos media de secundaria y de grado, para calcular la media de college y obtener un valor aproximado de los salarios de college
media_secundaria = round(df_union[df_union['Education'] == 'High School or Below']['Salary'].mean(),2)
```

```python
media_grado = round(df_union[df_union['Education'] == 'Bachelor']['Salary'].mean(),2)
```

```python
media_college = (media_secundaria + media_grado) / 2
media_college #media de college, estudios preuniversitarios: en base a secundaria y grado
```

```python
#filas que cumplan esta condición, creando un máscara booleana
mask = (df_union['Education'] == 'College') & (df_union['Salary'].isna()) 
```

```python
#pasamos la condición, la columna, y queremos que nos rellene los nan con la media de college que hemos obtenido previamente
df_union.loc[mask, 'Salary'] = media_college
```
